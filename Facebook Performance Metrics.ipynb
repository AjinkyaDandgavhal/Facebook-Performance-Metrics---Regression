```python
# Outlier Removal

plt.figure(figsize = (10,7))
df.boxplot()
plt.xticks(rotation = 90)
plt.show()

# There are outliers present in multiple columns. We will be using IQR method to remove the outliers.
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
UL = Q3 + 1.5*IQR
LL = Q1 - 1.5*IQR

df[((df > UL) | (df < LL)).any(axis = 1)].shape

# We ae having nearly 151 outliers. 
# We can cap them instead of removing them so that we do not lose out on valuable information.
# We will cap them at UL and LL.
```

```python
con_col = df.select_dtypes(include = np.number)
cat_col = df.select_dtypes(include = object)

for col in con_col.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    UL = Q3 + 1.5*IQR
    LL = Q1 - 1.5*IQR
    for ind in df[col].index:
        if df.loc[ind, col] > UL:
            df.loc[ind, col] = UL
        if df.loc[ind, col] < LL:
            df.loc[ind, col] = LL
        else:
            pass
        
```

```python
plt.figure(figsize = (10,7))
df.boxplot()
plt.xticks(rotation = 90)
plt.show()

# We have successfully capped all the outliers.
```

```python
# We need to treat multi-colineartiy before we fit the model.
# We can also check for multi-collinearity using Variance Inflation Factor.
```

```python
X = df.drop('Total Interactions', axis = 1) # Independent variables
y = df['Total Interactions'] # Target variable
```

```python
X1 = X.select_dtypes(include = np.number)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

v_f = pd.DataFrame()
v_f['Features'] = X1.columns
v_f['VIF'] = [vif(X1.values, i) for i in range(X1.shape[1])]
v_f.sort_values(by = 'VIF', ascending = False)

# We can drop the columns having VIF Score greater than 20.
```

```python
to_drop = v_f[v_f['VIF'] > 20]['Features'].tolist()

X = X.drop(to_drop, axis = 1)
```

```python
to_drop 
# We are dropping these columns.
```

```python
# We are considering only the variables which are having VIF Score < 8 for our model.
```

```python
# Encoding categorical columns

X = pd.get_dummies(X, drop_first = True)
```

```python
# Train-Test-Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 10, test_size = 0.2)
```

```python
# Linear Regression

import statsmodels.api as sm

linreg = sm.OLS(y_train, sm.add_constant(X_train)).fit()
print(linreg.summary())
```

```python
# From the above summary, we can infer that:
# 1. The R-Squared Value = 0.997, which is very high.
# 2. The prob(F-statistic) < 0.05 (alpha), therefore, the model is significant.
# 3. The Durbin-Watson Score = 2.047 ~ 2 indicating that auto-correlation is not present.
# 4. The Prob(JB) < 0.05 (alpha), indicating that the data is not normally distributed.
# 5. Condition No. > 1000, indicating presence of high multi-collinearity.
# 6. From the p-values for the different variables, we can understand which variables are significant and which are not.
# 7. We can build model using the significant variables only.
```

```python
p_val = pd.DataFrame(linreg.pvalues, columns = ['P_Value'])
imp_var = p_val[p_val['P_Value'] < 0.05].index.tolist()
imp_var
# Following are the significant variables.
```

```python
# Linear Regression

import statsmodels.api as sm

linreg = sm.OLS(y_train, X_train[imp_var]).fit()
print(linreg.summary())
```

```python
# From the above summary, we can infer that:
# 1. The R-Squared Value = 0.999, which is very high.
# 2. The prob(F-statistic) < 0.05 (alpha), therefore, the model is significant.
# 3. The Durbin-Watson Score = 2.028 ~ 2 indicating that auto-correlation is not present.
# 4. The Prob(JB) < 0.05 (alpha), indicating that the data is not normally distributed.
# 5. Condition No. > 1000, indicating presence of high multi-collinearity.
# 6. From the p-values for the different variables, we can observe that all the variables are significant.
# 7. For a unit change in the variable 'comment', the target variable changes by 1.1681 units.
# 8. For a unit change in the variable 'like', the target variable changes by 1.0586 units.
# 9. For a unit change in the variable 'share', the target variable changes by 0.7503 units.
# 10. For a unit change in the variable 'Type_Video', the target variable changes by 6.1053 units.
```

```python
y_pred_train = linreg.predict(X_train[imp_var])
y_pred_test = linreg.predict(X_test[imp_var])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
RMSE_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
RMSE_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
MAE_train = np.sqrt(mean_absolute_error(y_train, y_pred_train))
MAE_test = np.sqrt(mean_absolute_error(y_test, y_pred_test))
MAPE_train = np.sqrt(mean_absolute_percentage_error(y_train, y_pred_train))
MAPE_test = np.sqrt(mean_absolute_percentage_error(y_test, y_pred_test))
```

```python
print('RMSE_train :', RMSE_train)
print('RMSE_test :', RMSE_test)
print('MAE_train :', MAE_train)
print('MAE_test :', MAE_test)
print('MAPE_train :', MAPE_train)
print('MAPE_test :', MAPE_test)
```

```python
# We can observe that the values of RMSE (Root Mean Squared Error), MAE (Mean Absoulte Error), 
# MAPE (Mean Absoulte Percentage Error) for train and test set are very close.
# Therefore, we infer that there is no overfitting. 
```
